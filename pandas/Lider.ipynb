```python
import pandas as pd
import numpy as np
archivo= 'Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv'
punto_1= "Sea la función de costo C(x) = n*x^2 + 2*x + 23. \nHallar el costo promedio de producir de 100 a 105 unidades (Poner dos decimales)"
df= pd.read_csv(archivo)
def suma_digitos(num):
    return sum(int(d) for d in str(num) if d.isdigit())
df["n"] = df["Número de documento"].apply(suma_digitos)
def C(n, x):
    return n*(x**2) + 2*x + 23
valor_correcto = ((C(df["n"], 105) - C(df["n"], 100)) / 5).round(2)
df["CP1"] = (df.iloc[:,1] == valor_correcto).astype(int) * 4 + 1
print (df)
```
