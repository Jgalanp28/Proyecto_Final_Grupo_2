```python
import pandas as pd
import numpy as np
archivo = 'Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv'
Punto2 = "Sea la función de costo C(x) = n*x^2 + 2*x + 23./n Hallar el costo promedio de producir  de 100 a 101 unidades (Poner dos decimales)"
df = pd.read_csv(archivo)

def suma_digitos(num):
    return sum(int(d) for d in str(num) if d.isdigit())
def C (n,x):
    return n*(x**2)+2*x+23 
df["n"]= df["Número de documento"].apply(suma_digitos)

print("Calificar punto 2 ")

valor_correcto_p2 = (C(df["n"], 101) - C(df["n"], 100)).round(2)
df["CP2"]= (df.iloc[:,2]== valor_correcto_p2).astype(int)*4+1
print ("punto 2 calificado")
print(df[["Número de documento", "CP2"]].head())

print ("Calificando Punto 3 ")

columna_p3 = df.iloc[:,3]
def limpiar_valor(valor):
    try:
        if isinstance(valor, str):
            valor = valor.replace(",",".")
            valor = valor.strip()

            if "," in valor and "." in valor :
                partes = valor.split(",")
                if len(partes)== 2 : 
                    valor = partes[0]+"."+partes[1]

        return float(valor)
    except (ValueError,AttributeError):
        return np.nan
valores_limpios_p3 = columna_p3.apply(limpiar_valor)

valor_correcto_p3 = (2*df["n"]*100+2).round(2)
df["CP3"]=np.where(valores_limpios_p3 == valor_correcto_p3,5,1)
print("Punto 3 calificado")
print(df[["Número de documento", "CP3"]].head())
```
